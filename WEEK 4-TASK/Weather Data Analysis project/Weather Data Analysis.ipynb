```python
weather=df['Weather'].value_counts()
dfweather=pd.DataFrame(weather)
dfweather = dfweather.reset_index()
dfweather.columns = ['Weather', 'Frequency'] # change column names
dfweather
```

```python
#To rename column named 'Weather' to Weather Condition'
df.rename(columns = {'Weather' : 'Weather_Condition'}, inplace=False)
```

```python
df.groupby('Weather').get_group('Clear')
```

```python
#To find all records from data of when the weather was exactly clear
df[df['Weather'] == 'Clear']
```

```python
#To find the mean temperature, wind speed and visibility
print(df['Temp_C'].mean())
print(df['Wind Speed_km/h'].mean())
print(df['Visibility_km'].mean())
```

```python
#To find the variance of pressure
print(df['Press_kPa'].var())
```

```python
#To find the days on which wind speed was less than or equal to 30 km/hr and temperature was greater than 0 C
windspeed=df[(df['Wind Speed_km/h'] <=30) & (df['Temp_C']>0)]
print(windspeed['Date/Time'])
print(windspeed.shape)
```

```python
#To find the date and temperatures for all instances when snow was recorded
snow=df[df['Weather'] == 'Snow']
snow.loc[:, ['Date/Time', 'Temp_C']]
```

```python
#To display a graph of variation of temperature with respect to time
graph=df[['Temp_C', 'Date/Time']]
graph.plot(x='Date/Time', y='Temp_C',figsize=(10,7) )
```

```python
#To display a pie chart of percentages of weather conditions
#pie chart
keep=dfweather[dfweather['Frequency']>20]
print(keep)
keep.plot.pie(y='Frequency',autopct='%1.1f%%', shadow=False, figsize=(12,9))
plt.title("Weather Conditions")
plt.show()
```

```python
sample_data = df.head(120)

plt.figure(figsize=(10, 5))
plt.plot(sample_data['Date/Time'], sample_data['Temp_C'], color='orange', marker='.', linestyle='-')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trend - First 5 Days of Jan (Line Chart)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Convert 'Date/Time' from a string to a Datetime object
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
# ----------------

# 1. Bar Chart: Average Temperature by Month
df['Month'] = df['Date/Time'].dt.month_name()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_avg_temp = df.groupby('Month')['Temp_C'].mean().reindex(month_order).reset_index()

plt.bar(monthly_avg_temp['Month'], monthly_avg_temp['Temp_C'], color='skyblue')
plt.xlabel('Month')
plt.ylabel('Average Temperature ($^\circ$C)')
plt.title('Average Monthly Temperature')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
