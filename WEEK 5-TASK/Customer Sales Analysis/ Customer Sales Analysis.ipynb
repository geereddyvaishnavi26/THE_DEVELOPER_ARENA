##  Data Cleaning & Preparation - Handle missing values, convert dates, create calculated columns

**Handle Missing Values**

```python
# Drop rows that contain any missing values
df_cleaned = df.dropna()
```

```python
# Fill missing numerical values with column mean
df['MonthlyCharges'] = df['MonthlyCharges'].fillna(df['MonthlyCharges'].mean())
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
# Fill missing categorical values with column mode
df['PaymentMethod'] = df['PaymentMethod'].fillna(df['PaymentMethod'].mode()[0])
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df.dtypes
```

```python
df['Quantity'] = df['Quantity'].astype(int)
df['Price'] = df['Price'].astype(float)
```

```python
df['Calculated_Total_Sales'] = df['Quantity'] * df['Price']
```

```python
df['ARPU'] = df['TotalCharges'] / df['Tenure']
```

```python
df['Tenure_Years'] = df['Tenure'] / 12
```

```python
df['High_Value_Customer'] = df['TotalCharges'] > df['TotalCharges'].mean()
```

```python
# Check duplicates
df.duplicated().sum()

# Remove duplicates if any
df = df.drop_duplicates()
```

```python
df.info()
```

## Advanced Analysis - Create pivot tables, calculate retention rate, identify cross-selling

```python
pivot_sales = pd.pivot_table(
    df,
    values='TotalCharges',
    index='Region',
    columns='Product',
    aggfunc='sum',
    fill_value=0
)

pivot_sales
```

```python
# Pivot table to analyze monthly sales year-wise
pivot_monthly_sales = pd.pivot_table(
    df,
    values='TotalCharges',                 # Sales values
    index=df['Date'].dt.month_name(),      # Rows = Month names
    columns=df['Date'].dt.year,            # Columns = Year
    aggfunc='sum'                           # Sum of sales
)

pivot_monthly_sales
```

```python
# Pivot table to count unique customers per Region and Product
pivot_customers = pd.pivot_table(
    df,
    values='Customer_ID',     # Customer ID column
    index='Region',          # Rows = Region
    columns='Product',       # Columns = Product
    aggfunc='nunique'        # Count unique customers
)

pivot_customers
```

```python
# Convert Date column to Month-Year format
df['Month_Year'] = df['Date'].dt.to_period('M')
```

```python
# Count number of active customers in each month
monthly_customers = (
    df.groupby('Month_Year')['Customer_ID']
    .nunique()               # Count unique customers
)

monthly_customers
```

```python
# Count number of months each customer appears in
retained_customers = (
    df.groupby('Customer_ID')['Month_Year']
    .nunique()
)

retained_customers
```

```python
# Total unique customers
total_customers = df['Customer_ID'].nunique()

# Customers who appeared in more than one month (retained customers)
returning_customers = (retained_customers > 1).sum()

# Retention Rate formula
retention_rate = (returning_customers / total_customers) * 100

retention_rate
```

```python
# Create a table with unique Customer-Month combinations
customer_month = df[['Customer_ID', 'Month_Year']].drop_duplicates()
```

```python
# Merge table with itself to compare previous and current months
retention = customer_month.merge(
    customer_month,
    on='Customer_ID',
    suffixes=('_prev', '_curr')
)
```

```python
# Keep only forward month transitions
retention = retention[
    retention['Month_Year_curr'] > retention['Month_Year_prev']
]
```

```python
# Calculate month-wise retention rate
monthly_retention_rate = (
    retention.groupby('Month_Year_curr')['Customer_ID']
    .nunique() / monthly_customers.shift(1)
) * 100

monthly_retention_rate
```

```python
# Get list of products purchased by each customer
customer_products = (
    df.groupby('Customer_ID')['Product']
    .unique()
)

customer_products
```

```python
# Identify customers who purchased more than one product
multi_product_customers = customer_products[
    customer_products.apply(len) > 1
]

multi_product_customers.head()
```

```python
# Import required libraries for combination and counting
from itertools import combinations
from collections import Counter
```

```python
# Create empty list to store product pairs
product_pairs = []

# Loop through each customer's product list
for products in customer_products:
    # If customer bought more than one product
    if len(products) > 1:
        # Generate all product combinations (pairs)
        product_pairs.extend(combinations(sorted(products), 2))
```

```python
# Count frequency of each product pair
cross_sell_counts = Counter(product_pairs)

cross_sell_counts
```

```python
# Get top 10 most frequently bought product combinations
top_cross_selling = cross_sell_counts.most_common(10)

top_cross_selling
```
